```python
from pathlib import Path

import automol

geo_file = Path("opt.hess.v006.xyz")
geo, mode = automol.geom.from_xyz_string_with_mode(geo_file.read_text())
geo_forw = automol.geom.displace(geo, +0.1 * mode)
geo_back = automol.geom.displace(geo, -0.1 * mode)
automol.geom.display(geo)
automol.geom.display(geo_forw)
automol.geom.display(geo_back)

geo_forw_path = Path("path_init_forw.xyz")
geo_back_path = Path("path_init_back.xyz")
geo_forw_path.write_text(automol.geom.xyz_string(geo_forw))
geo_back_path.write_text(automol.geom.xyz_string(geo_back))
```
