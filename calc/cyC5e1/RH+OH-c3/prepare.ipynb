```python
import automol
from pathlib import Path

geo_start0_file = Path("neb_start_unsrt.xyz")
geo_end0_file = Path("neb_end_unsrt.xyz")

geo_start0 = automol.geom.from_xyz_string(geo_start0_file.read_text())
geo_end0 = automol.geom.from_xyz_string(geo_end0_file.read_text())

geo_start = automol.geom.align_to_atoms(geo_start0, (0, 1, 3), oop_idxs=(3, 6))
geo_end0 = automol.geom.align_to_atoms(geo_end0, (3, 4, 0), oop_idxs=(0, 6))

rh_end = automol.geom.subgeom(geo_start, tuple(range(13)))
oh_end = automol.geom.subgeom(geo_end0, (13, 14))
geo_end = rh_end + oh_end

geo_start_file = Path("neb_start.xyz")
geo_end_file = Path("neb_end.xyz")

geo_start_file.write_text(automol.geom.xyz_string(geo_start))
geo_end_file.write_text(automol.geom.xyz_string(geo_end))

automol.geom.display(geo_start, label=True)
automol.geom.display(geo_end, label=True)
```
